```python
# Create the target variable: Final inning score
inning_final_score = df.groupby(['match_id', 'inning'])['cumulative_runs'].max().reset_index()
```

```python
df = df.merge(inning_final_score, on=['match_id', 'inning'], suffixes=('', '_final'))
```

```python
# Drop unnecessary columns
drop_columns = [
    'match_id', 'batter', 'bowler', 'non_striker', 'id', 'city', 'date',
    'match_type', 'player_of_match', 'team1', 'team2', 'toss_winner', 'toss_decision',
    'winner', 'result', 'result_margin'
]
```

```python
df = df.drop(columns=drop_columns)
```

```python
# Encode categorical variables
categorical_columns = ['batting_team', 'bowling_team', 'match_phase', 'venue']
encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])
```
